```python
dataset = []
with open('iris.data') as file:
    for line in file:
        if line.strip() != "":
            cols = line.split(',')
            data = [float(cols[i]) for i in range(len(cols)-1)]
            data.append(cols[-1])
            dataset.append(data)
```

```python
print(len(dataset))
```

```python
import numpy as np
setosa_data = [data for data in dataset if "setosa" in data[-1]]
setosa_data = np.array(setosa_data)
print(setosa_data[:,0])
```

```python
versicolor_data = [data for data in dataset if "versicolor" in data[-1]]
versicolor_data = np.array(versicolor_data)
virginica_data = [data for data in dataset if "virginica" in data[-1]]
virginica_data = np.array(virginica_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(0,10,1)
Y = np.sin(X)
fig = plt.figure()
ax = plt.gca()
ax.grid()
ax.plot(X,Y, label = "y=sin(x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```
